**5.** M√¨nh quan t√¢m ƒë·∫øn c√°c app ti·∫øng anh, n√™n s·∫Ω l·ªçc c√°c app c√≥ t√™n b·∫±ng ti·∫øng anh.
- Vi·∫øt h√†m ki·ªÉm tra t√™n app: checkEnglishName(name).
- H√†m tr·∫£ v·ªÅ True n·∫øu t√™n app c√≥ t·ªëi ƒëa 3 k√≠ t·ª± ascii kh√¥ng n·∫±m trong kho·∫£ng (0-127).
- Ch·∫•p nh·∫≠n 3 k√≠ t·ª± n·∫±m ngo√†i kho·∫£ng 0-127 v√¨ c√≥ nh·ªØng app c√≥ t√™n g·ªìm k√≠ t·ª± emoji nh∆∞: "Instachat üòú" 
- Sau ƒë√≥ l·ªçc c√°c app c√≥ t√™n th·ªèa m√£n h√†m checkEnglishName, ( apply)

```python
def checkEnglishName(name):
  cnt = 0
  for i in name:
    if ord(i) < 0 or ord(i) > 127:
      cnt += 1
  if cnt <= 3:
    return True
  else:
    return False
app[app.App.apply(checkEnglishName)]
```

```python
app.Installs.unique()
```

```python
app.loc[app['Installs'] == 'Free']
```

```python
#app1 = app[app['Installs'] != 'Free']
def handleString(s):
  if s == 'Free':
    s = '0'
  elif s == '0':
    return s
  elif '+' in s:
    s = s.replace('+', '')
    if ',' in s:
      s = s.replace(',', '')
  return s
  
#handleString('12,000,000+')

app.Installs.apply(handleString).astype(int)

```
