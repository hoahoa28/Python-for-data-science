2. Viết hàm tính n! ( không dùng đệ quy)

```python
def tinhGT(n):
  gt = 1;
  for i in range(2, n + 1):
    gt *= i
  return gt
tinhGT(4)
```

3. Viết hàm tính số chữ hoa trong chuỗi.

```python
def sumUpCase(string):
  sum = 0
  for i in string:
    if 65 <= ord(i) <= 90:
      sum += 1
  return sum
sumUpCase('AhoaHBuT')


```

4. Viết hàm kiểm tra 1 số có phải số nguyên tố không ( hàm check chỉ được return True or False).

```python
import math
def checkPrime(n):
  if n == 1:
    return False
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True
checkPrime(199)
```

5. Viết hàm kiểm tra 1 số có phải số hoàn hảo không.

```python
import math
def checkPerfectNum(n):
  sum = 1
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      sum += i
      if i != n / i:
        sum += n / i
  #print(sum)
  if n == sum:
    return True
  else:
    return False
checkPerfectNum(9)
    

```

```python
def calSum(n):
 if n == 1:
  return 1
 else:
  return n + calSum(n - 1)
calSum(10)
```

7, In ra số fibonaci thứ n. (n>50)

```python
# def fiboNum(n):
#  if n == 1 or n == 2:
#    return 1
#  else:
#    return fiboNum(n - 1) + fiboNum(n - 2)
# fiboNum(51)

def fibonaci(n):
  if n == 1 or n == 2:
    return 1
  i = 3
  a1 = 1
  a2 = 1
  while(i <= n):
    a = a1 + a2
    a1 = a2
    a2 = a
    i += 1
  return a
fibonaci(50)
```

8. Cho list emails = ['buying books at amazom.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'] tìm các email valid trong list

```python
 # _@_.com
import re

valid = []
emails = ['buying books at amazom.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com']

pattern = r'\b[A-Za-z0-9]+@[A-Za-z0-9]+\.com\b' 

for i in emails: 
  if re.search(pattern, i):
    valid.append(i)

print(valid)
```

9,
Mật khẩu mạnh là mật khẩu thỏa mãn các tiêu chí sau:
• Độ dài >= 8
• Ít nhất có 1 chữ hoa và 1 chữ thường
• Ít nhất 1 ký tự đặc biệt (!@#$&*)
• Ít nhất có 1 số.
Cho list mật khẩu = ['123456' , 'Ntt#NT1210' , 'ai4E!python00', 'tumotden10']
tìm các mật khẩu mạnh.

```python
import re
pw = ['123456' , 'Ntt#NT1210' , 'ai4E!python00', 'tumotden10']
strongPW = []
pattern = "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!@#$&*])[A-Za-z\d!@#$&*]{8,}$"
for i in pw:
  if re.search(pattern, i):
    strongPW.append(i)
print(strongPW)

```
