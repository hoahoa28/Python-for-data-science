1. Input 1 chuỗi, output ra số kí tự hoa, số kí tự thường, số chữ số và số kí tự khác trong chuỗi.

```python
str1 = input('Nhap chuoi: ')
so = 0
chuThuong = 0
chuHoa = 0
khac = 0
for i in str1:
  if 48 <= ord(i) <= 57:
    so += 1
  elif 65 <= ord(i) <= 90:
    chuHoa += 1
  elif 97 <= ord(i) <= 122:
    chuThuong += 1
  else:
    khac += 1
print('So ky tu hoa la: ', chuHoa)
print('So ky tu thuong la: ', chuThuong)
print('So chu so la ', so)
print('So ky tu khac la ', khac)
```

2. Input 1 chuỗi, output ra 1 dict với key là ký tự, value là số lần chữ đấy xuất hiện. ví dụ: google.com-> {'o': 3, 'g': 2, '.': 1, 'e': 1, 'l': 1, 'm': 1, 'c': 1}

```python
str2 = input('Nhap chuoi: ')
myDict = {}
uniqueStr = set(str2)

for i in uniqueStr:
  myDict[i] = str2.count(i)

print(myDict)
```

3. Thay thế tất cả kí tự đầu tiên ( cả viết hoa cả viết thường) trong chuỗi bằng dấu $ (trừ kí tự đầu tiên) ví dụ: "Nguyen Thanh Tuan Nga" -> "Nguye$ Tha$h Tua$ $ga" ( kí tự đầu tiên là N, thay thế tất cả các kí tự N, n ở trong chuỗi ban đầu mà có vị trí khác 0 bằng dấu $)

```python
str3 = input('Nhap chuoi: ')
for i in range(1, len(str3)):
  if str3[i].lower() == str3[0].lower():
    str3 = str3[0: i] + '#' + str3[i+1 : len(str3)] 
print(str3)

```

4. Input động từ, output động từ thêm ing. Thêm theo các quy tắc sau
- Động từ tận cùng bằng e, bỏ e câm rồi thêm ing
- Động từ tận cùng bằng ie, ta biến ie thàng Y rồi thêm ing
- Nhân đôi phụ âm cuối trước khi thêm ing
- Còn lại chỉ cần thêm ing.

```python
str4 = input('Nhap dong tu: ')
size = len(str4)
if str4[size - 2 : size] == 'ie':
  str4 = str4[0 : size - 2] + 'y' + 'ing'
elif str4[size - 1] == 'e':
  str4 = str4[0 : size - 1] + 'ing'
elif (str4[-1] not in ['u', 'e', 'o', 'a', 'i']):
  str4 = str4[0 : size - 1] + str4[-1]*2 + 'ing'
else:
  str4 = str4 + 'ing'


print(str4)
```

6. Hàm replace trong python thay thế vị trí của tất cả các lần xuất hiện.
Cho người dùng nhập vào chuỗi
Cho người dùng nhập chuỗi con muốn thay thế
Cho người dùng nhập chuỗi được thay thế chuỗi con trên
Cho người dùng nhập vị trí thay thế (chỉ số bắt đầu từ 0)
Chỉ thay thế tại vị trí nhập, in ra chuỗi sau khi thay thế.
VD:
Input: "The apple the apple",
replace: "apple" -> "banana",
pos: 1 (index từ 0)
-> output: "The apple the banana"
( Mọi người tự viết lại hàm chứ không dùng luôn hàm replace nhé )

```python
string = input('Nhap chuoi: ')
oldSub = input('Nhap chuoi con duoc thay the: ')
newSub = input('Nhap chuoi con thay the: ')
pos = int(input('Nhap vi tri thay the: '))
count = 0
myDict = {}

for i in range( len(string) - len(oldSub) + 1):
  if string[ i : i + len(oldSub) ] == oldSub:
    count += 1
    myDict[count - 1] = i

print(myDict)

index = myDict[pos]
#if index == 0:
string = string[0:index] + newSub + string[index + len(oldSub) : len(string)]
print(string)


```
