8. Tại cuộc họp thường niên của Hội đồng quản trị , mọi người bắt đầu bắt tay với những người khác trong phòng. Thực tế là hai người bất kỳ bắt tay chính xác một lần, tính tổng số lần bắt tay?
Input : Nhập vào tổng số người
Output: In ra số lần bắt tay
Input: 1
Output: 0 Chỉ có 1 thành viên trong hội đồng , do đó 0 lần bắt tay.
Input : 2
Output: 1 Có 2 thành viên hội đồng quản trị, 1 lần bắt tay diễn ra.

```python
import math
from math import factorial

T = int(input('Number of testcases: '))
for i in range(1, T+1):
  print('Test case ' + str(i) + ':')
  n = int(input('Total people in the room: '))
  if (n == 1):
    totalHandShake = 0
  else:
    totalHandShake = int(factorial(n) / (factorial(2) * factorial(n - 2)))
  print('Total handshakes are: ' + str(totalHandShake) + '\n')
```
