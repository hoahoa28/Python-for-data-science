a. Tính số lượng sản phẩm trong dataset.

b. Lấy ra các sản phẩm có giá lớn hơn 50000.

c. Lấy ra các sản phẩm đã có doanh thu hơn 300000.

```python
import numpy as np

dataset = np.array([
['Quần thun', 10, 50000],
['Áo gió', 5, 90000],
['Tất', 50, 8000],
['Giày', 1, 200000],
['Mũ', 1, 30000],
['Quần jean', 3, 100000],
['Áo len', 10, 100000],
])

#a
c2 = dataset[:, 1].astype(int)
print(f'Tong so luong san pham la: {np.sum(c2)}')

#b
mask = dataset[:, 2].astype(int) > 50000
sp = dataset[mask]
print(f'Cac san pham co gia lon hon 50000 la: {sp[:, 0]}')

#c
doanhThu = dataset[:, 1].astype(int) * dataset[:, 2].astype(int)
#print(doanhThu)

ds = np.column_stack([dataset[:, 0], doanhThu])
#print(ds)

mask2 = ds[:, 1].astype(int) > 300000
res = ds[mask2]
print(f'Cac san pham co doanh thu lon hon 300000 la {res[:, 0]} ')
```

```python
import numpy as np

n = int(input('Nhap so hang va so cot: '))
print('Nhap cac phan tu cua ma tran (cach nhau boi dau cach):')
entries = list(map(int, input().split()))
#print(entries)
#matrix = np.reshape(entries, (n, n))
matrix = np.array(entries).reshape(n, n)
print('Ma tran ban dau la:')
print(matrix)

def live(matrix):
  nextMatrix = np.zeros_like(matrix)
  for i in range(n):
    for j in range(n):
      check = np.sum(matrix[i - 1 : i + 2, j - 1 : j + 2]) - matrix[i][j]
      if matrix[i][j] == 1 and check < 2:
        nextMatrix[i][j] = 0
      elif matrix[i][j] == 1 and (check == 2 or check == 3):
        nextMatrix[i][j] = 1
      elif matrix[i][j] == 1 and check > 3:
        nextMatrix[i][j] = 0
      elif matrix[i][j] == 0 and check == 3:
        nextMatrix[i][j] = 1
      elif matrix[i][j] == 0:
        nextMatrix[i][j] = 0

  return nextMatrix

print('The he tiep theo cua ma tran la:')
print(live(matrix))

```
