6) Nhập vào a,b,c. Giải và biện luận tìm nghiệm phương trình ax^2 + bx + c = 0. (a = 0 vẫn thỏa mãn vì không bắt buộc là phương trình bậc 2).

```python
import math
a = float(input('a = '))
b = float(input('b = '))
c = float(input('c = '))
if a == 0:
  if b == 0:
    if c == 0:
     print('Phuong trinh vo so nghiem')
    else:
     print('Phuong trinh vo nghiem')
  else:
    x = -c/b
    print(f'Phuong trinh co nghiem x = {x}')
else:
  delta = b*b - 4*a*c
  if delta > 0:
    x1 = (-b - math.sqrt(delta)) / (2*a)
    x2 = (-b + math.sqrt(delta)) / (2*a)
    print(f'Phuong trinh co 2 nghiem, x1 = {x1}, x2 = {x2}')
  elif delta == 0:
    x = -b/(2*a)
    print(f'Phuong trinh co nghiem kep x1 = x2 = {x}')
  else:
    print('Phuong trinh vo nghiem')

```
