1. In ra các chữ cái từ a đến z.



```python
character = 'a'
d = ord(character)
while(d <= ord('z')):
  print(chr(d), end =' ')
  d += 1
```

2. Nhập vào 1 số n, in ra tổng các số từ 1 đến n. Ví dụ: n=10, in ra 55.

```python
n = int(input('n = '))
tong = 0
for i in range(n+1):
  tong += i
print('tong:',tong)
```

4. Nhập vào 1 số n, in ra các ước của n và nhỏ hơn n, ví dụ n = 6, in ra 1, 2, 3

```python
n = int(input('n = '))
i = 1
while (i <= n/2):
  if n % i == 0:
    print(i, end = ' ')
  i += 1
```

6. Số hoàn hảo là số là nó bằng tổng các ước của nó, ví dụ 6 = 1 + 2 + 3. In ra các số hoàn hảo lớn hơn 2 và nhỏ hơn 1000.

```python
for n in range(3, 1000):
  i = 1
  sum = 0
  while i <= n/2:
    if n % i == 0:
      sum += i
    i += 1
  if sum == n:
    print(n, end = ' ')
```

8.
“Vừa gà vừa chó
Bó lại cho tròn
Ba mươi sáu(36) con
Một trăm(100) chân chẵn
Hỏi mấy gà, mấy chó?”
Giải bài toán trên dùng vòng lặp.

```python
for i in range(1,37):
  for j in range(1,37-i):
    if (i + j == 36) and (i * 2 + j * 4 == 100):
      print(f"so ga la {i}, so cho la {j}")
```

```python
for i in range(36):
  if i * 2 + (36 - i) * 4 == 100:
    print(f'so ga la {i}, so cho la {36 - i}')
```

9. Nhập vào n > 0. Tính S = 1/2 + 2/3 + 3/4 + ... + n/(n + 1)

```python
n = int(input('n = '))
S = 0
if (n > 0):
  for i in range(1, n+1):
    S += i/(i + 1)
print('S = ',S)
```
