(2đ) Một của hàng bán áo phông với giá 20$ 1 cái. Mức độ giảm giá phụ
thuộc vào số lượng áo mua như sau:
Số lượng Mức giảm giá

1-9 0%

10-19 10%

20-39 25%

>40 40%

Cho người dùng nhập số lượng áo muốn mua, tính ra thành tiền


```python
n = int(input('Nhap so luong ao muon mua: '))
thanhTien = 0
if 1 <= n <= 9:
  thanhTien = n * 20
elif 10 <= n <= 19:
  thanhTien = (n * 20) * 0.9
elif 20 <= n <= 39:
  thanhTien = (n * 20 ) * 0.75
elif n >= 40:
  thanhTien = (n * 20) * 0.6
print('So tien phai tra la: ', thanhTien)
```

2. (4đ) Viết 1 chương trình cho quản lý sinh viên, cho phép giáo viên nhập
điểm từ bàn phím:
"Tên học sinh? "
"Điểm? "

Cho đến khi giáo viên nhập tên học sinh là END thì dừng. Giả thiết là
giáo viên nhập điểm trong khoảng [0-10]

(a) (2đ) Lưu thông tin học sinh và điểm dưới dạng dict, nếu học sinh
đã có trong dict thì in thông báo hỏi người dùng có muốn cập nhật
không? .

(b) (1đ) Tính số lượng học sinh giỏi [8-10], khá [6.5-8), trung bình [5-6.5),
yếu trong lớp.

(c) (1đ) In ra tên của 3 bạn cao điểm nhất trong lớp, nếu điểm trùng
xét theo thứ tự bảng chữ cái.


```python
#a
qlSV = {}
while True:
  ten = (input('Tên học sinh ? '))
  if ten == 'end':
    break
  else:
    if ten in qlSV:
      traLoi = input('Bạn có muốn cập nhật điểm không: ')
      if traLoi == 'có':
        qlSV[ten] = float(input('Điểm ? '))
      if traLoi == 'không':
        continue
    else:
      qlSV[ten] = int(input('Điểm ? '))
print(qlSV)

#b
gioi = 0
kha = 0
tb = 0
yeu = 0
for i in qlSV.values():
  if 8 <= i <= 10:
    gioi += 1
  elif 6.5 <= i < 8:
    kha += 1
  elif 5 <= i < 6.5:
    tb += 1
  elif 0 <= i < 5:
    yeu += 1

print('Số lượng học sinh giỏi là: ', gioi)
print('Số lượng học sinh khá là: ', kha)
print('Số lượng học sinh trung bình là: ', tb)
print('Số lượng học sinh yếu là: ', yeu)

#c
dsDiem = list(qlSV.values())
top3 = []
dsDiem.sort(reverse=True)
dsDiem = dsDiem[ : 3]
for i in dsDiem:
  for j in qlSV.keys():
    if qlSV[j] == i:
      if j in top3:
        continue
      else:
        top3.append(j)
print(top3)





```

(b) (1đ) Việt hàm kiểm tra xem 1 số có tính chất sau hay không: nó là
số nguyên tố và đảo ngược của số đấy cũng là số nguyên tố. Ví dụ:
13 là số nguyên tố, 31 cũng là số nguyên tố. Nhưng 19 là số nguyên
tố còn 91 = 7 * 13, không phải là số nguyên tố.

(c) (1đ) In ra các số nguyên dương nhỏ hơn 1000 thỏa mãn tính chất
phần b.

```python
#a
n = int(input('Nhap n: '))
def daoNguoc(n):
  soDaoNguoc = 0
  while(n > 0):
    a = n % 10
    soDaoNguoc = soDaoNguoc * 10 + a
    n = n // 10
  return soDaoNguoc
daoNguoc(n)
```

```python
#b
import math
n = int(input('Nhap n: '))
def check(n):
  if n == 1:
    return False
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  for i in range(2, int(math.sqrt(daoNguoc(n))) + 1):
    if daoNguoc(n) % i == 0:
      return False
  return True
check(n)
```

```python
#c
for i in range(1, 1000):
  if check(i):
    print(i)
```
